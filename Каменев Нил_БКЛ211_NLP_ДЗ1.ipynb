```python
#привести к нижнему регистру и начальной форме
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
digits = [str(i) for i in range(10)]
def preprocess(tokens1):
    return [morph.normal_forms(word1)[0]
            for word1 in tokens1
                if (word1[0] not in digits)]
def preprocess2(tokens2):
    return [morph.normal_forms(word2)[0]
            for word2 in tokens2
                if (word2[0] not in digits)]
```

```python
#составить 2 множества — слова только в положительных и только в отрицательных отзывах
for e in preprocess(tokens1):
    if e in preprocess(tokens2):
        preprocess(tokens1).remove(e)
        preprocess(tokens2).remove(e)
positive_set = set(preprocess(tokens1))
negative_set = set(preprocess(tokens2))
```

```python
#фильтр по частотности
from collections import Counter
def collect_freqlist(negative_set):
    freqlist = Counter()
    for wrod in negative_set:
        if wrod.isalpha():
            freqlist[wrod] += 1
    return dict(freqlist.most_common())
```

```python
#создать функцию — предсказывание — отзыв положительный или отрицательный (каких отличительных слов больше)
text_input = input()
def prognoz(text_input):
    positive_count = 0
    negative_count = 0
    for t in nltk.word_tokenize(text_input):
        if t in positive_set:
            positive_count += 1
        elif t in negative_set:
            negative_count += 1
    if positive_count >= negative_count:
        return "Отзыв положительный!"
    else:
        return "Отзыв отрицательный("
prognoz(text_input)
```
