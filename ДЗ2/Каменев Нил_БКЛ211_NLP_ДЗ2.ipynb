```python
with open("Мои планы.txt", "r", encoding="utf-8") as f:
    text = f.read()
```

```python
# pymorphy
from pymorphy2 import MorphAnalyzer
m = MorphAnalyzer()
from pymorphy2.tokenizers import simple_word_tokenize
for token in simple_word_tokenize(text):
    print(m.parse(token))
```

#### Этот текст подходит для оценки из-за некоторых трудных и неоднозначных для автоматического POS-теггинга моментов — так, здесь есть слова с дефисом в написании (бизнес-английский), некоторые имена собственные (Юрием Виктровичем), случаи омонимии (права)

#### Для разметки использовались POS-теги тегсета Universal Dependencies (17 основных)
